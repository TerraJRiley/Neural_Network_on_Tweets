```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Natural Language Processing libraries, initiations and functions
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from nltk.stem import WordNetLemmatizer
from nltk import word_tokenize
from nltk.corpus import stopwords
from sklearn.decomposition import LatentDirichletAllocation
import re # Delete this if scraping in same notebook

def preprocess(text):
    text = re.sub(r'[^a-zA-Z]',' ', text.lower())
    tokens = word_tokenize(text)
    lemmer = WordNetLemmatizer()
    stop_words = stopwords.words("english")
    return " ".join([lemmer.lemmatize(word) for word 
                     in tokens if len(word) > 1 and not word in stop_words])
cvec = CountVectorizer(analyzer = "word",
                       min_df = 2,
                       preprocessor = preprocess,
                       stop_words = 'english')
#pd.DataFrame(cvec.fit_transform(df['COLUMN']).todense(), columns=cvec.get_feature_names())

def LDA(df_column, num_topics = 3, num_words = 5):
    # Cols are the words. Rows are the topics
    topic_lists = []
    lda = LatentDirichletAllocation(n_components=num_topics, learning_method='online')
    lda.fit(cvec.fit_transform(df_column))
    for ix, topic in enumerate(lda.components_):
        topic_lists += [[cvec.get_feature_names()[i] for i 
                         in lda.components_[ix].argsort()[:-num_words - 1:-1]]]

    return pd.DataFrame(topic_lists, columns=[ 'word_' + str(i) for i 
                                    in range(1, num_words+1)], index=range(1, num_topics + 1)) 
```

## Modeling

```python
X = df[features]
y = df['Target']
X_train, X_test, y_train, y_test = train_test_split(X,y)
ss = StandardScaler()
X_train_scaled = ss.fit_transform(X_train)
X_test_scaled = ss.transform(X_test)
```

```python
# Ensure to stratify the train/test when you do that, for this model type
model = Sequential()                                # Note: Start here.  Grow & add after 1st model
model.add(Dense(150,                                # Start == to # of features, but not over 100
          activation = 'relu',                      # Default.  Rarely need to change
          input_dim=X_train.shape[1]))              # Keeps consistent without hasstle, not needed post-1st layer

model.add(Dense(140, activation = 'relu')) # Here, have another layer
model.add(Dense(130, activation = 'relu')) # Here, have another layer
model.add(Dense(120, activation = 'relu')) # Here, have another layer
model.add(Dense(110, activation = 'relu')) # Here, have another layer
model.add(Dense(100, activation = 'relu')) # Here, have another layer
model.add(Dense(90, activation = 'relu')) # Here, have another layer
model.add(Dense(80, activation = 'relu')) # Here, have another layer
model.add(Dense(70, activation = 'relu')) # Here, have another layer
model.add(Dense(60, activation = 'relu')) # Here, have another layer
model.add(Dense(50, activation = 'relu')) # Here, have another layer
model.add(Dense(40, activation = 'relu')) # Here, have another layer
model.add(Dense(30, activation = 'relu')) # Here, have another layer
model.add(Dense(20, activation = 'relu')) # Here, have another layer
model.add(Dense(10, activation = 'relu')) # Here, have another layer
model.add(Dense(5, activation = 'relu')) # Here, have another layer

#model.add(Dense(neurons_for_this_layer, activation = 'relu')) # Here, have another layer
model.add(Dense(1, activation='sigmoid')) # Output Layer.  Not to be changed from this for Binary Classification
model.compile(optimizer='adam',
              loss='binary_crossentropy',           # Loss function for Binary Classification
             metrics=['accuracy']                  # Optional metric
             )

results = model.fit(X_train_scaled, y_train,
                    epochs=25,                      # Default is 1, but start with about 10
                    batch_size = 32,               # Unsure if useable in this model, but whatever
                    #verbose=2,                     # This will silence printing output
                    validation_data=(X_test_scaled, y_test))
```

```python
# Visualisations for classification Models
# Plots to view Loss/Accuracy over epoche's
plt.plot(results.history['val_loss'], label='Test Loss')
plt.plot(results.history['loss'], label='Train Loss')
plt.legend();
plt.plot(results.history['val_acc'], label='Test Accuracy')
plt.plot(results.history['acc'], label='Train Accuracy')
plt.legend();

#Alternative
fig, ax = plt.subplots(ncols=2, nrows=1, figsize=(15,4))

ax[0].plot(results.history['loss'])
ax[0].set_title("Loss")
ax[0].set_xlabel("epochs")

ax[1].plot(results.history['acc'])
ax[1].set_title("Accuracy")
ax[1].set_xlabel("epochs")
```
